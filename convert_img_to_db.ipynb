```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

# make readable txt report of dates for fileobjects in a dfxml file
import os
import argparse
import sys
import re
import pandas as pd
import hashlib
# import Objects.py from python dfxml tools
#sys.path.append('/Users/sunflower/PHD/python')
#sys.path.append(os.path.join(os.path.dirname(__file__), ".."))
import dfxml
import dfxml.objects as objects2
# import objects2
json_path =  "/Users/seunfuta/Downloads/M57/charlie-2009-12-11.json"
img_path =  "/Users/seunfuta/Downloads/M57/charlie-2009-12-11.E01"
dfxml_path =  "/Users/seunfuta/Downloads/M57/charlie-2009-12-11.xml"

num_files = int()
num_files = 0
out_json = open(json_path, 'w')
with open(img_path, 'rb') as img_h:
    # gather info for each FileObject
    filestr_list = []
    for (event, obj) in objects2.iterparse(dfxml_path):
        # Only work on FileObjects
        if not isinstance(obj, objects2.FileObject):
            continue

        #app = args.app
        filemd5 = obj.md5
        filepath = obj.filename
        filesize = obj.filesize
        byte_runs = obj.byte_runs
        if filemd5 and not re.search(r'\.$', filepath):
            num_files += 1
            #{"file_hash": "54aa35cbe9d1fb383cbc33e1879b210e","file_type": "","filesize": 45966800,"name_pairs": ["Chrome28-W7x64","Users/user/AppData/Local/Google/Chrome/Application/28.0.1500.95/chrome.dll"],"nonprobative_count": 9800,"zero_count": 779}
            filestr = "{\"file_hash\": \"" + filemd5+ "\",\"file_type\": \"\",\"filesize\":"+str(filesize)+",\"name_pairs\": [\"app\",\""+filepath+"\"],\"nonprobative_count\": 0,\"zero_count\": 0}"
            if filestr not in filestr_list: filestr_list.append(filestr)
            # print("app", app, "filemd5", filemd5, "filepath", filepath, "filesize", filesize, "byte_runs", type(byte_runs))
            for runs in obj.byte_runs.__iter__():
                print("runs", runs.img_offset, runs.len, runs.file_offset, runs.fs_offset)
                if runs.img_offset and runs.len:
                    img_h.seek(runs.img_offset)
                    block_hash=""
                    for block in range(0,int(float(runs.len)/float(512))+1):
                        block_hash = hashlib.md5(img_h.read(512)).hexdigest()
                        strx = "{\"block_hash\":\""+block_hash+"\",\"img_offset\":"+str(int(runs.img_offset)+int(int(block) * int(512)))+",\"source_sub_counts\":[\""+filemd5+"\",1,["+str(int(runs.file_offset+(512*block)))+"]],\"block_label\":\"XX\"}"
                #print("filemd5", filemd5, "filepath", filepath, "filesize", filesize, "byte_runs",byte_runs, type(byte_runs.__iter__()), type(byte_runs.__repr__()))
                        print(strx)
                        out_json.write(strx)
                        out_json.write("\n")
                else:
                    continue
    for filestr in filestr_list:
        print(filestr)
        out_json.write(filestr)
        out_json.write("\n")
```
